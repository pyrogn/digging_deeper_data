```python
import pandas as pd
import numpy as np
import scipy.stats
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('cw.csv')
```

```python
df.head(1)
```

```python
df.iloc[:, 1].value_counts()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
df.iloc[:, 4].value_counts().plot(kind='bar', color='black')
plt.grid(alpha=0.3)
plt.yticks(np.arange(0, 12, 1))
plt.title('Кол-во курсовых на научного руководителя')
plt.ylabel('Кол-во курсовых')
plt.show()
fig.savefig('1.png', bbox_inches='tight', dpi=700)
```
